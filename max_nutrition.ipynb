# Exploring the MaxNutritionEnv

```python
from stable_baselines3 import PPO, DQN, A2C
from meal_planning_environment import MaxNutritionEnv
from meal_planning_environment import run_with_learning_algorithm, load_data

import os 
import time
from IPython.display import clear_output
```

## Load all meal, nutrition, and carbon data

```python
possible_meals, meal_categories, nutrition_data, carbon_data = load_data()
clear_output()
```

## Initialize MealPlanningEnv
We can choose from any of the following environments:
1. MaxNutritionEnv
2. HeuristicEnv
3. GPTOnlyEnv
4. RLHFEnv

In this notebook we will expore the MaxNutritionEnv and design a meal-plan for one week (i.e., 21 meals).

```python
num_meals = 21
max_env = MaxNutritionEnv(
    possible_meals=possible_meals, 
    meal_categories=meal_categories, 
    nutrition_data=nutrition_data, 
    carbon_data=carbon_data, 
    num_meals=num_meals
)
```

# MaxNutritionEnv with PPO vs. DQN vs. A2C

We'll compare each of our different learning algorithms for the chosen environment.

```python
log_dir = f"./tmp/gym/{time.strftime('%Y%m%d-%H%M%S')}"
num_steps = 100000
log_dir += '_max_env_PPO' + f',num_steps={num_steps}'
os.makedirs(log_dir, exist_ok=False)
print(f'Output stored at: {log_dir}')
run_with_learning_algorithm(PPO, max_env, num_steps, log_dir, num_meals=21, seed=123)
```

```python
log_dir = f"./tmp/gym/{time.strftime('%Y%m%d-%H%M%S')}"
num_steps = 100000
log_dir += '_max_env_DQN' + f',num_steps={num_steps}'
os.makedirs(log_dir, exist_ok=False)
print(f'Output stored at: {log_dir}')
run_with_learning_algorithm(DQN, max_env, num_steps, log_dir, num_meals=21, seed=123)
```

```python
log_dir = f"./tmp/gym/{time.strftime('%Y%m%d-%H%M%S')}"
num_steps = 100000
log_dir += '_max_env_A2C' + f',num_steps={num_steps}'
os.makedirs(log_dir, exist_ok=False)
print(f'Output stored at: {log_dir}')
run_with_learning_algorithm(A2C, max_env, num_steps, log_dir, num_meals=21, seed=123)
```
